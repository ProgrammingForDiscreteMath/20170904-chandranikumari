```python
def expSumApprox(n):
    """Return the function to approximate the exponential function"""
    f(x)=0
    for i in range(n):
        f(x)=f(x)+(x^i/factorial(i))
    return (f(x))
a=animate([plot(expSumApprox(k),xmin=-3,xmax=3,ymin=0,ymax=20) for k in range (10)])
a.show()
```
